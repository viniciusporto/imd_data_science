```python
import numpy as np

def calcStep():
    dice = np.random.randint(6) + 1
    if dice in [1,2]:
        return -1
    elif dice in [3,4,5]:
        return 1
    elif dice == 6:
        return np.random.randint(6) + 1
        


games = []

for g in range(500):
    round = []
    floor = 0
    for r in range(100):
        step = calcStep()
        w = np.random.random(1)
        if w < 0.001:
            step = 0
        floor += step    
        floor = max(0, floor)
        round.append(floor)
    games.append(round)
#print(games)
ganhador = 0
for idx,r in enumerate(games):
    print('Jogador: %d | Final Stair: %d' % (idx+1,int(r[-1])))

pontuacao_final = [row[-1] for row in games];

print('**************************************************')
print('Jogador Vencedor: %d | Final Stair: %d' % (pontuacao_final.index(max(pontuacao_final)),int(max(pontuacao_final))))
    
```
